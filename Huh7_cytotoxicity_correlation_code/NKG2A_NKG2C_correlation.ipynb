Donor 10- Donor 17, Each donor has 6 data conditions 1-6
Donor D32-D34, beta4,beta7,beta10,beta 11, cond2, cond3



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the data
data = np.loadtxt('absolute_MFI_cyto_receptor.dat')

# Extract receptor expressions
NKG2C = data[:, 10]  # Column index 11 in MATLAB is 10 in Python (0-based indexing)
NKG2A = data[:, 11]  # Column index 12 in MATLAB is 11 in Python

#print((NKG2C[0:5]))

 #Calculate the correlation between NKG2A and NKG2C
correlation, p_value = pearsonr(NKG2A, NKG2C)
print(f'Correlation between NKG2A and NKG2C: {correlation:.2f}, p-value: {p_value:.5f}')
# Set the font to Arial
plt.rcParams['font.family'] = 'Arial'

# Create a scatter plot
#plt.figure(figsize=(8, 6))
plt.scatter(NKG2A, NKG2C, alpha=0.7, edgecolor='k', s=100)
plt.xlabel('NKG2A Expression', fontsize=12, fontweight='bold')
plt.ylabel('NKG2C Expression', fontsize=12, fontweight='bold')
plt.title('Scatter Plot of NKG2A vs. NKG2C', fontsize=14, fontweight='bold')
#plt.grid(True)

# # Optionally, add a trend line
# p = np.polyfit(NKG2A, NKG2C, 1)  # Linear fit
# yfit = np.polyval(p, NKG2A)
# plt.plot(NKG2A, yfit, '-r', linewidth=2, label='Trend Line')

# Customize the plot
#plt.legend(loc='best')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
